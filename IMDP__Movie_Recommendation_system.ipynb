```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data collection and preprocessing

```python
movies_data = pd.read_csv('/content/movieMetadata (1).csv')
```

```python
movies_data.head()
```

```python
# find the rows and columns in tje data feame
movies_data.shape
```

```python
movies_data.reset_index(inplace=True)
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# selecting the relevent features of the recomandation
selected_features = ['Description','Genre','Certificate','Director','Stars']
print(selected_features)
```

```python
# replacing the null values with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 seleted features
combined_features = movies_data['Description']+' '+movies_data['Genre']+' '+movies_data['Certificate']+' '+movies_data['Director']+' '+movies_data['Stars']
print(combined_features)
```

```python
# converting the text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectorizer = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectorizer)
```

Cosine Similarity

```python
# getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectorizer)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movieSimilarity = pd.DataFrame(similarity, columns= movies_data['Movie Title'], index=movies_data['Movie Title']) # convert to dataframe
movieSimilarity.head()
```

```python
# getting the movie name the user
movie_name = input('Enter your favorite movie name: ')
```

```python
# ctreating a list with all the movie names given in the dataset
list_of_all_titles =movies_data['Movie Title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match )
```

```python
# finding the index of the movie with title
index_of_the_movie = movies_data[movies_data['Movie Title'] == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on there similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda X:X[1],reverse=True)
```

```python
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index
print('Movies suggested for you \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['Movie Title'].values[0]
  if(i<40):
    print(i,'.',title_from_index)
    i += 1
```

Movie Recomandation System

```python
# Predictive system
movie_name = input('Enter your favorite movie name: ')
list_of_all_titles =movies_data['Movie Title'].tolist()
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
close_match = find_close_match[0]
index_of_the_movie = movies_data[movies_data['Movie Title'] == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda X:X[1],reverse=True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['Movie Title'].values[0]
  if(i<40):
    print(i,'.',title_from_index)
    i += 1

```
