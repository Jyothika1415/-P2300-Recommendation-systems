```python
# Calculate average Metascore score for each genre
avg_metacore = df.groupby('Genre')['Metascore'].mean().reset_index()

# Get the top 10 genres by average Metascore
top_10_genres = avg_metacore.nlargest(25, 'Metascore')

# Plotting a barplot for the top 10 genres
plt.barh(top_10_genres['Genre'], top_10_genres['Metascore'],color = 'g')
plt.xlabel('Average Metascore Score')
plt.ylabel('Genre')
plt.title('Top 10 Genres by Average Metascore')
plt.show()
```

```python
# Visualize how the average rating or Metascore has changed over the years.
# Are newer movies generally better

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Drop rows with NaN values in 'Year', 'Rating', and 'Metascore' columns
df = data.dropna(subset=['Year', 'Metascore'])

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Group by 'Year' and calculate the average rating and Metascore for each year
avg_scores_by_year = df.groupby('Year')[['Metascore']].mean()

# Line plot for Average Rating and Metascore over the years
plt.figure(figsize=(12, 4))
sns.lineplot(data=avg_scores_by_year, markers=True, palette="tab10")
plt.title('Average Metascore Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.legend(['Metascore'])
plt.show()
```

```python
# Drop rows with NaN values in 'Year', 'Rating', and 'Metascore' columns
df = data.dropna(subset=['Year', 'Rating'])

# Convert 'Year' to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Group by 'Year' and calculate the average rating and Metascore for each year
avg_scores_by_year = df.groupby('Year')[['Rating']].mean()

# Line plot for Average Rating and Metascore over the years
plt.figure(figsize=(10,4))
sns.lineplot(data=avg_scores_by_year, markers=True, palette="tab10")
plt.title('Average Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.legend(['Rating'])
plt.show()
```

```python
df = data.dropna(subset=['Certificate'])

# Count the occurrences of each certificate type
certificate_counts = df['Certificate'].value_counts()

# Bar plot for distribution of movies across certificate types
plt.figure(figsize=(10, 4))
sns.barplot(x=certificate_counts.index, y=certificate_counts.values, palette="Set3")
plt.title('Distribution of Movies Across Certificate Types')
plt.xlabel('Certificate Type')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Calculate average rating for each certificate
avg_ratings = df.groupby('Certificate')['Rating'].mean().reset_index()

# Plotting a barplot
plt.subplots(figsize=(4,7))
plt.barh(avg_ratings['Certificate'], avg_ratings['Rating'], color='plum')
plt.xlabel('Average Rating')
plt.ylabel('Certificate')
plt.title('Average Rating by Certificate')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = data.dropna(subset=['Gross_Income', 'Rating'])

# Clean 'Gross_Income' column by removing non-numeric characters
df['Gross_Income'] = df['Gross_Income'].replace('[^\d.]', '', regex=True)

# Convert 'Gross_Income' to numeric
df['Gross_Income'] = pd.to_numeric(df['Gross_Income'], errors='coerce')

# Scatter plot for Gross Income vs. User Ratings
plt.figure(figsize=(10, 4))
sns.scatterplot(x='Gross_Income', y='Rating', data=df, color='skyblue', alpha=0.7)
plt.title('Gross Income vs. User Ratings')
plt.xlabel('Gross Income')
plt.ylabel('User Rating')
plt.show()

# Calculate the correlation coefficient
correlation_coefficient = df['Gross_Income'].corr(df['Rating'])
print(f"Correlation Coefficient: {correlation_coefficient}")
```
